```python
class Dice:
    def roll(self):
        return random.randint(1, 6)
    
class Player:
    def __init__(self, name):
        self.name = name
        self.score = 0
        
    def play_turn(self, dice: Dice):
        roll_value = dice.roll()
        self.score += roll_value
        print(f"{self.name} rolled a {roll_value}, total score: {self.score}")
        
class Game:
    def __init__(self, player1, player2, rounds=5):
        self.player1 = player1
        self.player2 = player2
        self.dice = Dice()
        self.rounds = rounds
        
    def start(self):
        for round_num in range(1, self.rounds + 1):
            print(f"\n ------Round {round_num} ---- ")
            self.player1.play_turn(self.dice)
            self.player2.play_turn(self.dice)
            
        print("--- Final Result ---")
        if self.player1.score > self.player2.score:
                print(f"{self.player1.name} wins the game!")
        elif self.player2.score > self.player1.score:
            print(f"{self.player2.name} wins the game!")
        else:
            print("It's a tie!")
            
game = Game(Player("Arun"), Player("Akshat"), rounds=5)
game.start()
            
```

```python
import random

class Player:
    CHOICES = ["rock", "paper", "scissor"]
    
    def __init__(self, name, is_human=False):
        self.name = name
        self.score = 0
        self.is_human = is_human
    
    def make_choice(self):
        if self.is_human:
            choice = input(f"{self.name}, enter rock/paper/scissor: ").lower()
        else:
            choice = random.choice(Player.CHOICES)
        print(f"{self.name} chose {choice}")
        return choice

class Game:
    def __init__(self, player1, player2, rounds=3):
        self.player1 = player1
        self.player2 = player2
        self.rounds = rounds
    
    def decide_winner(self, choice1, choice2):
        if choice1 == choice2:
            return None
        win_rules = {"rock": "scissor", "scissor": "paper", "paper": "rock"}
        return self.player1 if win_rules[choice1] == choice2 else self.player2
    
    def start(self):
        for round_num in range(1, self.rounds + 1):
            print(f"\n--- Round {round_num} ---")
            choice1 = self.player1.make_choice()
            choice2 = self.player2.make_choice()
            
            winner = self.decide_winner(choice1, choice2)
            if winner:
                winner.score += 1
                print(f"{winner.name} wins this round!")
            else:
                print("This round is a tie!")

            print(f"Score Update: {self.player1.name}={self.player1.score}, {self.player2.name}={self.player2.score}")


        print("\n=== Final Result ===")
        if self.player1.score > self.player2.score:
            print(f"{self.player1.name} is the overall winner!")
        elif self.player2.score > self.player1.score:
            print(f"{self.player2.name} is the overall winner!")
        else:
            print("The game is a draw!")


game = Game(Player("Alice",is_human=True), Player("Computer"), rounds=3)
game.start()
```


##### Two players (batsmen) play a 6-ball(1 over) cricket match each.

- Each ball can score random runs (0â€“6).

- Keep track of runs and balls faced for each player.

- After all turns, print the scoreboard and the winner.

- Use classes: Player, Game.

- Add mid updates after each over.

```python
class Balling:
    def runs(self):
        return random.randint(0, 6)

class Player:
    def __init__(self, name):
        self.name = name
        self.score = 0
    
    def batting(self, ball: Balling):
        get_run = ball.runs()
        self.score += get_run
        print(f"{self.name} scored {get_run} in this ball, total runs: {self.score}")
        
class Game:
    def __init__(self, player1, player2, over=1):
        self.player1 = player1
        self.player2 = player2
        self.ball = Balling()
        self.over = over
        
    def start(self):
        for ball in range(1, (self.over * 6) + 1):
            print(f"{self.player2.name} {ball} ball: ")
            self.player1.batting(self.ball) 
        
        print("\n ")
        
        for ball in range(1, (self.over * 6) + 1):
            print(f"{self.player1.name} {ball} ball: ")
            self.player2.batting(self.ball)
            
        
        print(" ----- Final Result ----")
        if self.player1.score > self.player2.score:
            print(f"{self.player1.name} = {self.player1.score}, {self.player2.name} = {self.player2.score}")
            print(f"{self.player1.name} wins by {self.player1.score - self.player2.score} runs")
        elif self.player1.score < self.player2.score:
            print(f"{self.player1.name} = {self.player1.score}, {self.player2.name} = {self.player2.score}")
            print(f"{self.player2.name} wins by {self.player2.score - self.player1.score} runs")
        else:
            print("Match Tie")
            
cricket = Game(Player("Raghav"), Player("Kamaal"), over=1)
cricket.start()
```
